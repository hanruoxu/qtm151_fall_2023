(a) Compute function for each individual (map)

<font size = "3">

- Import the datset "data/bmi.csv"
    - **Note that the column `Height` is in $cm$, and `Weight` is in $kg$**
- Write a lambda function for $ BMI = \frac{weight (kg)}{height (m)^2} $
- Use ```list(map(...))``` to calculate $BMI$ for each row of the dataset
- Round the outcome into 2 decimal points
- Store the calculation in a new column in the dataframe called `BMI`
- Display the dataframe


````python
dataset = pd.read_csv("data/bmi.csv")
BMI = lambda weight, height_cm: round(weight/(height_cm/100)**2, 2)
dataset['BMI']=list(map(lambda weight, height:BMI(weight, height), dataset['Weight'], dataset['Height']))
dataset
````

(b) Convert to upper case

- Create the following function to convert a string to upper case

$\quad$ ``` fn_upper = lambda x: x.upper() ```

- Create a new column, "countryname_upper"  <br>
that converts "countryname" to upper case.

- Print the original `countryname` and the newly created column, like: <br>
"['Afghanistan', 'Africa Eastern and Southern', 'Africa Western and Central', 'Albania', 'Algeria', ......] "

````python
# Write your own code here
fn_upper = lambda countryname: countryname.upper()
dataset = pd.read_csv("data/wdi_2020.csv")
dataset['countryname_upper'] = dataset['countryname'].apply(fn_upper)
countryname_list = list(dataset['countryname'])
countryname_upper_list = list(dataset['countryname_upper'])
print(countryname_list)
print(countryname_upper_list)
````

(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here
dataset = pd.read_csv("data/wdi_urban.csv")
list_unique_year = pd.unique(dataset["year"])

for year in list_unique_year:
    data_for_year = dataset[dataset["year"] == year]
    plt.hist(data_for_year["prop_urbanpopulation"], alpha=0.5, label=str(year))

plt.xlabel("Proportion of Urban Population")
plt.ylabel("Frequency")
plt.title("Urban Population Distribution by Year")
plt.legend(list_unique_year)
````
